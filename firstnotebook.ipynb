```python
prep = pandas.DataFrame( test.loc[ (test["amount"] == 287.08 ) ] )
print(prep.shape)
print( 'Total of 287.08 contributions: ', prep['amount'].sum() )
print( "Total support contributions: ", test['amount'].sum() )
print( 'Percent 287.08 of total support: ', prep['amount'].sum()/test['amount'].sum()  )

prep.head(5)
```
